```python
import numpy as np
import matplotlib.pyplot as plt

sensors = ["Temperature", "Battery", "Distance"]
values = [92, 45, 15]

plt.bar(sensors, values, color=["red", "green", "blue"])
plt.title("Robot Sensor Status")
plt.ylabel("Value")
plt.grid(True)
plt.show
```

```python

import numpy as np
import matplotlib.pyplot as plt

speed = np.array([5, 10, 15, 20, 25, 30])
distance = np.array([2, 8, 18, 32, 50, 72])

plt.scatter(speed, distance, color="red", s=100)
plt.title("Speed vs Distance")
plt.xlabel("Speed (km/h)")
plt.ylabel("Distance (m)")
plt.grid(True)
plt.show()
```
