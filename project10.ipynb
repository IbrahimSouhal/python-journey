```python
import matplotlib.pyplot as plt
import numpy as np

def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

def sort_battery_descending(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] < arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

temperature = [72, 45, 88, 31, 95, 60, 78, 52]
battery = [80, 65, 90, 45, 30, 55, 70, 20]
distance = [15, 42, 8, 67, 23, 51, 12, 38]

print("Temperature", bubble_sort(temperature))
print("Battery", sort_battery_descending(battery))
print("Distance", bubble_sort(distance))

print("Max Temperature:", np.max(np.array(temperature)))
print("Max Battery:", np.max(np.array(battery)))
print("Max Distance:", np.max(np.array(distance)))

print("Min Temperature:", np.min(np.array(temperature)))
print("Min Battery:", np.min(np.array(battery)))
print("Min Distance:", np.min(np.array(distance)))

print("Average Temperature:", np.mean(np.array(temperature)))
print("Average Battery:", np.mean(np.array(battery)))
print("Average Distance:", np.mean(np.array(distance)))

plt.plot(distance, battery, label="Battery")
plt.plot(distance, temperature, label="Temperature")

plt.title("Robot Sensor Data Analysis")
plt.grid(True)
plt.xlabel("Distance (km)")
plt.ylabel("Value")
plt.legend()
plt.show()
```
