```python
import numpy as np

temperatures = np.array([45, 78, 92, 55, 81, 60])

print("Temperatures:", temperatures)
print("Max:", np.max(temperatures))
print("Min:", np.min(temperatures))
print("Average:", np.mean(temperatures))
```

```python
battery = np.array([80, 45, 20, 95, 60])
distance = np.array([25, 5, 50, 12, 8])

print("Battery average:", np.mean(battery))
print("Closest obstacle:", np.min(distance))
print("Dangerous distances:", distance[distance < 10])
```

```python
time = np.array([1, 2, 3, 4, 5])
speed = np.array([0, 10, 25, 18, 5])

print("Time:", time)
print("Speed:", speed)
print("Max speed:", np.max(speed))
print("Average speed:", np.mean(speed))
print("Distance traveled", np.sum(speed))
print("Stopped moments", time[speed==0])
print("Fast moments", time[speed > 20])
```
