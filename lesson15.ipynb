```python
import matplotlib.pyplot as plt
import numpy as np

time = np.array([1, 2, 3, 4, 5])
speed = np.array([0, 10, 25, 18, 5])

plt.plot(time, speed)
plt.title("Robot Speed Over Time")
plt.xlabel("Time (s)")
plt.ylabel("Speed (km/h)")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

time = np.array([1, 2, 3, 4, 5])
temperature = np.array([45, 78, 92, 55, 60])
battery = np.array([80, 70, 55, 40, 20])
distance = np.array([30, 25, 8, 15, 5])

plt.plot(time, temperature, label="Temperature")
plt.plot(time, battery, label="Battery")
plt.plot(time, distance, label="Distance")

plt.title("Robot Sensors Over Time")
plt.xlabel("Time (s)")
plt. ylabel("Value")
plt.legend()
plt.grid(True)
plt.show()
```
